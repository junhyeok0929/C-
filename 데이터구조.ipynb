```python
import random
import time

def selection_sort_recursive(arr):
    # 재귀적 선택 정렬
    if len(arr) <= 1:
        return arr

    min_idx = 0
    for i in range(1, len(arr)):
        if arr[i] < arr[min_idx]:
            min_idx = i

    arr[0], arr[min_idx] = arr[min_idx], arr[0]
    return [arr[0]] + selection_sort_recursive(arr[1:])
```

```python
# 0에서 100까지의 자연수 300개를 가지는 리스트 생성
arr = [random.randint(0, 100) for _ in range(300)]

# 정렬 시간 측정
start_time = time.time()
selection_sort_recursive(arr)
end_time = time.time()

# 정렬 시간 출력
sort_time = end_time - start_time
print(f"Selection Sort (Recursive) 정렬 시간: {sort_time:.6f} seconds")
```

```python
import random
import time
import matplotlib.pyplot as plt

def selection_sort_recursive(arr):
    # 재귀적 선택 정렬
    if len(arr) <= 1:
        return arr

    min_idx = 0
    for i in range(1, len(arr)):
        if arr[i] < arr[min_idx]:
            min_idx = i

    arr[0], arr[min_idx] = arr[min_idx], arr[0]
    return [arr[0]] + selection_sort_recursive(arr[1:])

# 시간 측정을 위한 리스트 크기들
sizes = [100, 200, 300, 400, 500]

# 각 리스트 크기에 대한 정렬 시간 저장
times = []

# 정렬 시간 측정 및 저장
for size in sizes:
    # 0에서 100까지의 자연수를 크기에 맞게 생성
    arr = [random.randint(0, 100) for _ in range(size)]

    # 정렬 시간 측정
    start_time = time.time()
    selection_sort_recursive(arr)
    end_time = time.time()

    # 정렬 시간 저장
    sort_time = end_time - start_time
    times.append(sort_time)

# 그래프 그리기
plt.plot(sizes, times)
plt.xlabel('List Size')
plt.ylabel('Sorting Time (seconds)')
plt.title('Selection Sort (Recursive) Sorting Time')
plt.show()
```
